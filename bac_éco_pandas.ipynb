## Les méthodes à utiliser avec Pandas pour analyser Coronavirus_Tunisia.csv
J'ai acquis ces données de Mosaïque FM et de covid-19.tn.
Cet ensemble de données contient le statut du virus COVID-19 à 2020 en Tunisie.

```python
#récupérer les données collectées à partir de extension csv
#charger la bibliothéque pandas
from pandas import *             
#ou 2éme méthode: import pandas as pd
```

```python
#créer dataframe sous le nom corona, à partir de ficher Coronavirus_Tunisia.csv
corona=read_csv('Coronavirus_Tunisia.csv', header=0)
#afficher le 5 premier ligne de dataframe
print(corona.head())
```

```python
#Affichage du nombre d'éléments d'un dataFrame
print(corona.size)
```

```python
#Affichage des informations concernant un DataFrame
print(corona.info())
```

```python
#Affichage des dimensions d'un dataframe (nombre des lignes, nombre des colonnes)
print(corona.shape)
```

```python
#Affichage du résumé rapide d'un Dataframe (sur les colonnes de type numériques)
print(corona.describe())
```

```python
#mean() : valeurs moyennes
#min() : valeur minimale
#max() : valeur maximale
#sum() : calculer la somme des valeurs
#count() : nombre d’éléments d’un DataFrame
```

```python
#Afficher le contenu de la colonne 'date'
print(corona['date'])
```

```python
#modifier les données de la ligne d'indice 2 [09/03/2020,54,Mahdia, Paris, 22/02/2020,F,Critical,Sousse,University Hospital Farhat Hashad,high blood pressure]
corona.loc[2]=['09/03/2020',54,'Mahdia', 'Paris', '22/02/2020','F','Critical','Sousse','University Hospital Farhat Hashad','high blood pressure']
print(corona.head())
```

```python
#renommer la colonne "hospital_name" par "nom_hôpital"
corona.rename(columns={'hospital_name':'nom_hôpital'}, inplace=True)
print(corona.head())
```

```python
#Afficher la plus grand age
print(corona['age'].max())
```

```python
#afficher le nombre d'occurence de city
print(corona['city'].value_counts())
```

```python
#afficher les ages supérieur à 40
print(corona['age']>40)
```

```python
#afficher les city de tunis
print(corona['city']=='Tunis')
```

```python
#ordonner la dataframe par age dans l'ordre corissant
corona=corona.sort_values(by=['age'] , ascending=True)
print(corona)
```

## Géneration du graph 

```python
#charger la bibliothéque graphique Matplotlibe
import matplotlib.pyplot as plt
```

```python
#Création d’un graphique à barres : Colonne_Abscisse x est les noms des villes  
#le colonne_d'ordonner y est le colonne age 
corona.plot.bar(x='city',y='age',title='les ages des status',color='red')
```

```python
#Création d’un graphique en courbe :
#afficher la courbe d'age en fonction des vielles tunisienne
corona.plot.line(x='city', y='age', title='la courbe des ages', color='black')
```

```python
#autre bibliothéque graphique nommée seaborn
#charger la bib seaborn
import seaborn as sns
#redefine la taille de figure
plt.figure(figsize=(14, 8))
#en fonction de colonne city visualiser le nombre d'occurrence des villes 
sns.countplot(x='city', data=corona)
#Affichage d’un graphique
plt.show()
```
